```python
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

# 弱相關

```python
x = np.random.randint(0,50,1000)
y = np.random.randint(0,50,1000)
```

```python
np.corrcoef(x,y)
```

```python
plt.scatter(x,y)
```

# 正相關

```python
x = np.random.randint(0,50,100)
y = 2*x+np.random.randint(0,40,100)
np.corrcoef(x,y)
```

```python
plt.scatter(x,y)
```

# 負相關

```python
x = np.random.randint(0,50,100)
y = -5*x + np.random.randint(0,20,100) 
np.corrcoef(x,y)
```

```python
plt.scatter(x,y)
```
