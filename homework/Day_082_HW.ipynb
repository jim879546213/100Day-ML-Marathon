```python
import keras
import tensorflow as tf
gpu_options = tf.GPUOptions(allow_growth=True)
sess = tf.Session(config=tf.ConfigProto(gpu_options=gpu_options))
tf.keras.backend.set_session(sess)
```

```python
(x_train,y_train),(x_test,y_test) = keras.datasets.cifar10.load_data()
```

```python
def preproc_x(x, flatten=True):
    x = x/255
    if flatten:
        x = x.reshape(len(x),-1)
    return x
def preproc_y(y, num_classes=10):
    if y.shape[-1]==1:
        y = keras.utils.to_categorical(y, num_classes)
    return y
```

```python
x_train = preproc_x(x_train)
x_test = preproc_x(x_test)

y_train = preproc_y(y_train)
y_test = preproc_y(y_test)
```

```python
from keras.layers import Dropout,Dense, Input

def build_mlp(input_shape, output_units=10, num_neurons=[512,256,128], drop_ratio=0.2):
    input_layer = Input(input_shape)
    
    for i, num_unit in enumerate(num_neurons):
        if i == 0:
            x = Dense(units=num_unit, activation='relu', name='hidden_layer'+str(i+1))(input_layer)
            x = Dropout(drop_ratio)(x)
        else:
            x= Dense(units=num_unit, activation='relu', name='hidden_layer'+str(i+1))(x)
            x = Dropout(drop_ratio)(x)
    out =  Dense(units=output_units, activation='softmax', name='output')(x)
    
    model = keras.models.Model(inputs=[input_layer], outputs=[out])
    
    return model
```

```python
## 超參數設定
LEARNING_RATE = 1e-3
EPOCHS = 50
BATCH_SIZE = 256
MOMENTUM = 0.95
Dropout_EXP = [0.1,0.2,0.3,0.4]
```

```python
result = {}
for drp in Dropout_EXP:
    model = build_mlp(input_shape=x_train.shape[1:], drop_ratio=drp)
    model.summary()
    model.compile(loss="categorical_crossentropy", metrics=["accuracy"], optimizer='adam')
    model.fit(x_train, y_train, 
          epochs=EPOCHS, 
          batch_size=BATCH_SIZE, 
          validation_data=(x_test, y_test), 
          shuffle=True)
    train_loss = model.history.history["loss"]
    valid_loss = model.history.history["val_loss"]
    train_acc = model.history.history["acc"]
    valid_acc = model.history.history["val_acc"]
    drop_name = 'drop rate:'+str(drp)
    result[drop_name] = {'train-loss':train_loss,
                         'valid-loss':valid_loss,
                         'train-acc':train_acc,
                         'valid-acc':valid_acc}
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

color_bar =['r','g','b','k']
plt.figure(figsize=(6,8))
for i, cond in enumerate(result.keys()):
    plt.plot(range(len(result[cond]['train-loss'])),result[cond]['train-loss'], '-', c=color_bar[i], label=cond)
    plt.plot(range(len(result[cond]['valid-loss'])),result[cond]['valid-loss'], '--', c=color_bar[i], label=cond)
plt.legend()
plt.title('Loss')
plt.show()

plt.figure(figsize=(6,8))
for i, cond in enumerate(result.keys()):
    plt.plot(range(len(result[cond]['train-acc'])), result[cond]['train-acc'],'-', c=color_bar[i], label=cond)
    plt.plot(range(len(result[cond]['valid-acc'])), result[cond]['valid-acc'],'--', c=color_bar[i], label=cond)
plt.legend()
plt.title('Accuracy')
plt.show()
```
