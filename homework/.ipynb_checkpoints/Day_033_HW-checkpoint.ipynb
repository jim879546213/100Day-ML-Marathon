# 1. 模型的泛化能力 (generalization) 是指什麼？

Generalization refers to your model's ability to adapt properly to new, previously unseen data, drawn from the same distribution as the one used to create the model.

# 2. 分類問題與回歸問題分別可用的目標函數有哪些？

回歸問題:(https://buzzorange.com/techorange/2018/06/22/computer-learning-5-tips/)


MSE(L2):賦予異常點更大的權重，有異常點，就會備受干擾；MSE 損失的梯度隨損失增大而增大，而損失趨於 0 時則會減小


MAE(L1):更新的梯度始終相同，即使對於很小的損失值，梯度也很大，這樣不利於模型的學習(可以使用變化的學習率改善)


Huber:Huber 損失是絕對誤差，只是在誤差很小時(delta = 界線)，就變為平方誤差(缺點:Huber 損失的問題是我們可能需要不斷調整超參數 delta)


Log-Cosh 損失:比 L2 更平滑的的損失函數。它的計算方式是預測誤差的雙曲餘弦的對數
