Label encoding 

    Label encoding 的表示方式會讓同一個欄位底下的類別之間有大小關係 (0<1<2<...)

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

逐項練習

```python
#根據NAME_CONTRACT_TYPE去做encode
le.fit(app_train['NAME_CONTRACT_TYPE'])
```

```python
#看被encode的項目有甚麼
le.classes_
```

```python
#將被fit的le用在'NAME_CONTRACT_TYPE'這行內
le.transform(app_train['NAME_CONTRACT_TYPE'])
```

```python
#transfer的還原動作
le.inverse_transform([0,1,0,1,1,0])
```

```python
#創造新行去裝encoded 'NAME_CONTRACT_TYPE'
app_train['ENCODE_NAME_CONTRACT_TYPE']=le.transform(app_train['NAME_CONTRACT_TYPE'])
```

改變object數量小於兩項的行

```python
#datafram默認取值為行的名稱
for a in app_train:
    print(a)
    print(app_train[a].dtype)
    
```

```python
#le = LabelEncoder()   (上面已經定義過了) 
le_count = 0
le_name =[]
#le_mean =[]

for col in app_train:
    if app_train[col].dtype == "object":
        if len(list(app_train[col].unique()))<=2:
            le.fit(app_train[col])
            app_train[col]=le.transform(app_train[col])
            app_test[col]=le.transform(app_test[col])
            
            #紀錄有多少個col被encode
            le_count+=1
            #紀錄是誰被encode
            le_name.append(col)
            #紀錄各col code所代表的意義
            #le_mean[col]=list(le.inverse_transform[0,1]):'method' object is not subscriptable????
            #le_mean+=list(le.inverse_transform[0,1])
print('%d cols have been encoded'%le_count)
print(le_name)
#print(le_mean)
```

```python
#個別行
NAME_TYPE_SUITE_ENCODE_train = pd.get_dummies(app_train["NAME_TYPE_SUITE"])
NAME_TYPE_SUITE_ENCODE_test = pd.get_dummies(app_test["NAME_TYPE_SUITE"])

for i in b:
    print(NAME_TYPE_SUITE_ENCODE_train[i].head())
```

```python
NAME_TYPE_SUITE_ENCODE_train
```

```python
#合併
app_train = pd.concat([app_train,NAME_TYPE_SUITE_ENCODE_train],axis = 1)
```
