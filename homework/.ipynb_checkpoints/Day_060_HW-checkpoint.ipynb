```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.linear_model import SGDClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import warnings
warnings.filterwarnings("ignore")
```

```python
# 定義 PCA 與隨後的邏輯斯迴歸函數
log = SGDClassifier(loss='log', penalty='l2', max_iter=10000,tol=1e-5, random_state=0)
pca = PCA()
pipe = Pipeline(steps = [('pca',pca),('logistic',log)])

#載入MNIST
digits = datasets.load_digits()
X_digits = digits.data
y_digits = digits.target
```

```python
# 繪製不同 components 的 PCA 解釋度
pca.fit(X_digits)

fig, (ax0,ax1) = plt.subplots(nrows = 2, sharex =True, figsize = (6,6))
#sharex = True :x刻度一樣
ax0.plot(pca.explained_variance_ratio_, linewidth = 2)
ax0.set_ylabel('PCA explained variance')

ax0.axvline(search.best_estimator_.named_steps['pca'].n_components, linestyle = ':', label = 'n_components chosen')
#.axvline : Add a vertical line across the axes.
ax0.legend(prop = dict(size = 12))


# 繪製不同採樣點的分類正確率
results = pd.DataFrame(search.cv_results_)

components_cols = 'param_pca__n_components'
best_clfs = results.groupby(components_cols).apply(lambda g : g.nlargest(1,'mean_test_score'))
#df.nlargest:從df中根據指定的那行('mean_test_score')選出最大的n(n =1)行數據

best_clfs.plot(x = components_cols, y = 'mean_test_score', yerr = 'std_test_score', legend = False, ax = ax1)
#yerr : y的標準差(正負那個數值)ps還有其他設定
#ax = ax1 :指出這個圖要畫在哪個subplot裡面

ax1.set_ylabel('Classification accuracy')
ax1.set_xlabel('n_components')
plt.tight_layout()
#tight_layout会自动调整子图参数，使之填充整个图像区域
plt.show()
```
